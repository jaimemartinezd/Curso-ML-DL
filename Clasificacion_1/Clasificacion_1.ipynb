```python
import pandas as pd
from sklearn.model_selection import train_test_split
db = pd.read_csv('clasificacion1/train.csv')
db_test = pd.read_csv('clasificacion1/test.csv')
db_test
```

```python
from sklearn.preprocessing import LabelEncoder
X = db.drop(columns=['satisfaction', 'Arrival Delay in Minutes'])
X_prueba = db_test.drop(columns=['satisfaction', 'Arrival Delay in Minutes'])
y = db['satisfaction']
le = LabelEncoder()
X['Gender'] = le.fit_transform(X['Gender'])
X_prueba['Gender'] = le.transform(X_prueba['Gender'])
X['Customer Type'] = le.fit_transform(X['Customer Type'])
X_prueba['Customer Type'] = le.transform(X_prueba['Customer Type'])
X['Type of Travel'] = le.fit_transform(X['Type of Travel'])
X_prueba['Type of Travel'] = le.transform(X_prueba['Type of Travel'])
X['Class'] = le.fit_transform(X['Class'])
X_prueba['Class'] = le.transform(X_prueba['Class'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4).fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy', splitter='random', max_depth=3, random_state=43)
clf = clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, knn.predict(X_prueba))
print(f'Accuracy: {accuracy:.2f}')
```

```python
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(y_test, clf.predict(X_prueba)))
```

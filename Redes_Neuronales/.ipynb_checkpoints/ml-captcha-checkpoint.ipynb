```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from PIL import Image

# Cargar CSVs
train_df = pd.read_csv('captcha/train.csv')
val_df = pd.read_csv('captcha/validation.csv')

# Mostrar las primeras filas
train_df.head()
```

```python
# Mostrar 5 ejemplos del conjunto de entrenamiento
fig, axes = plt.subplots(1, 5, figsize=(15, 4))

for i, ax in enumerate(axes):
    row = train_df.iloc[i]
    img_id = f"{int(row['Id']):05d}"  # Formato 00000, 00001, ...
    img_path = os.path.join('captcha/train', f"{img_id}.png")
    image = Image.open(img_path)
    ax.imshow(image, cmap='gray')
    ax.set_title(f"Label: {row['Label']}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Parmetros
IMG_WIDTH = 200
IMG_HEIGHT = 50
MAX_LABEL_LEN = 6
CHARS = '0123456789'

# Mapeo carcter <-> ndice
char_to_idx = {char: i for i, char in enumerate(CHARS)}
idx_to_char = {i: char for char, i in char_to_idx.items()}

def load_and_preprocess_image(img_path):
    image = Image.open(img_path).convert('L')  # Escala de grises
    image = image.resize((IMG_WIDTH, IMG_HEIGHT))
    image_array = np.array(image) / 255.0       # Normalizar
    return image_array
```

```python
def encode_label(label):
    label = str(label).zfill(MAX_LABEL_LEN)  # Asegura que tenga 6 caracteres, con ceros delante si es necesario
    encoded = np.zeros((MAX_LABEL_LEN, len(CHARS)))
    for i, char in enumerate(label):
        encoded[i, char_to_idx[char]] = 1.0
    return encoded

# Por ahora, cargamos por ejemplo 5000 imgenes
N = 5000
X = np.zeros((N, IMG_HEIGHT, IMG_WIDTH, 1))
y = np.zeros((N, MAX_LABEL_LEN, len(CHARS)))

for i in range(N):
    row = train_df.iloc[i]
    img_id = f"{int(row['Id']):05d}"
    img_path = os.path.join('captcha/train', f"{img_id}.png")
    img = load_and_preprocess_image(img_path)
    X[i, :, :, 0] = img
    y[i] = encode_label(row['Label'])
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def build_captcha_model():
    input_img = layers.Input(shape=(IMG_HEIGHT, IMG_WIDTH, 1))

    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)

    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)

    x = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)

    x = layers.Flatten()(x)
    x = layers.Dense(512, activation='relu')(x)

    outputs = []
    for _ in range(MAX_LABEL_LEN):
        outputs.append(layers.Dense(len(CHARS), activation='softmax')(x))

    model = models.Model(inputs=input_img, outputs=outputs)
    return model

model = build_captcha_model()
model.summary()
```

```python
# Dividir las etiquetas en 6 arrays, uno por dgito
y_split = [y[:, i, :] for i in range(MAX_LABEL_LEN)]
```

```python
model.compile(
    loss=['categorical_crossentropy'] * MAX_LABEL_LEN,
    optimizer='adam',
    metrics=['accuracy'] * MAX_LABEL_LEN
)
```

```python
history = model.fit(
    X, y_split,
    batch_size=64,
    epochs=10,
    validation_split=0.1
)
```

```python
# Cargar y procesar el conjunto de validacin
val_df = pd.read_csv('captcha/validation.csv')

X_val = np.zeros((len(val_df), IMG_HEIGHT, IMG_WIDTH, 1))
y_val = np.zeros((len(val_df), MAX_LABEL_LEN, len(CHARS)))

for i, row in enumerate(val_df.itertuples()):
    img_id = f"{int(row.Id):05d}"
    img_path = os.path.join('captcha/validation', f"{img_id}.png")
    img = load_and_preprocess_image(img_path)
    X_val[i, :, :, 0] = img
    y_val[i] = encode_label(row.Label)
```

```python
# Evaluar el modelo en el conjunto de validacin (separando las salidas por dgitos)
val_metrics = model.evaluate(X_val, [y_val[:, i, :] for i in range(MAX_LABEL_LEN)], verbose=1)

# Mostrar las mtricas
for i in range(MAX_LABEL_LEN):
    print(f"Loss for digit {i + 1}: {val_metrics[i]}")
    print(f"Accuracy for digit {i + 1}: {val_metrics[i + MAX_LABEL_LEN]}")
```

```python
# Cargar las imgenes del conjunto de test
test_images = []
test_ids = []

# Recorrer todas las imgenes en la carpeta test
for img_name in os.listdir('captcha/test'):
    if img_name.endswith('.png'):
        img_id = int(img_name.split('.')[0])
        img_path = os.path.join('captcha/test', img_name)
        img = load_and_preprocess_image(img_path)
        test_images.append(img)
        test_ids.append(img_id)

# Convertir la lista de imgenes a un array numpy
X_test = np.array(test_images)
```

```python
# Funcin para decodificar las predicciones (probabilidades) en su dgito correspondiente
def decode_label(pred_probs):
    return str(np.argmax(pred_probs))  # Tomamos el ndice con la mayor probabilidad (0-9)

# Realizar las predicciones con el modelo
predictions = model.predict(X_test)

# Convertir predictions a un array de NumPy (si no lo es ya)
predictions = np.array(predictions)

# Ver las dimensiones de las predicciones
print(predictions.shape)  # Esto debera tener forma (n_imagenes, MAX_LABEL_LEN, 10)

# Decodificar las predicciones para cada imagen
decoded_predictions = []
for i in range(predictions.shape[0]):
    pred_digits = ''.join([decode_label(predictions[i, j]) for j in range(MAX_LABEL_LEN)])
    decoded_predictions.append(pred_digits)

# Mostrar las primeras 5 predicciones como ejemplo
for i in range(5):
    print(f"Id: {test_ids[i]}, Predicted Label: {decoded_predictions[i]}")
```

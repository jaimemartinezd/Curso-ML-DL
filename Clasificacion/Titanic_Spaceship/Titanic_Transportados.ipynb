```python
import pandas as pd
db_train = pd.read_csv('data/train.csv')
db_test = pd.read_csv('data/test.csv')
db_test
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

oh = OneHotEncoder(sparse_output=False)

X = db_train.drop(columns=['PassengerId', 'Name', 'Transported'])
X_prueba = db_test.drop(columns=['Id','PassengerId', 'Name'])
y = db_train['Transported']

cat_cols = ['HomePlanet', 'CryoSleep', 'VIP', 'Destination']

X_encoded = oh.fit_transform(X[cat_cols])  
encoded_df = pd.DataFrame(X_encoded, columns=oh.get_feature_names_out(cat_cols))

X = X.drop(columns=cat_cols).reset_index(drop=True)
X = pd.concat([X, encoded_df], axis=1)

X_prueba_encoded = oh.transform(X_prueba[cat_cols])
encoded_df_prueba = pd.DataFrame(X_prueba_encoded, columns=oh.get_feature_names_out(cat_cols))

X_prueba = X_prueba.drop(columns=cat_cols).reset_index(drop=True)
X_prueba = pd.concat([X_prueba, encoded_df_prueba], axis=1)

cabin_split = X['Cabin'].str.split('/', expand=True)
cabin_split.columns = ['Cabin_Letter', 'Cabin_Number', 'Cabin_Category']
cabin_split['Cabin_Number'] = pd.to_numeric(cabin_split['Cabin_Number'], errors='coerce')

X = pd.concat([X.drop(columns=['Cabin']), cabin_split], axis=1)

cabin_split_prueba = X_prueba['Cabin'].str.split('/', expand=True)
cabin_split_prueba.columns = ['Cabin_Letter', 'Cabin_Number', 'Cabin_Category']
cabin_split_prueba['Cabin_Number'] = pd.to_numeric(cabin_split_prueba['Cabin_Number'], errors='coerce')

X_prueba = pd.concat([X_prueba.drop(columns=['Cabin']), cabin_split_prueba], axis=1)

categorical_cols = ['Cabin_Letter', 'Cabin_Category']

X_encoded_cabin = oh.fit_transform(X[categorical_cols])
encoded_df_cabin = pd.DataFrame(X_encoded_cabin, columns=oh.get_feature_names_out(categorical_cols))

X = X.drop(columns=categorical_cols).reset_index(drop=True)
X = pd.concat([X, encoded_df_cabin], axis=1)

X_prueba_encoded_cabin = oh.transform(X_prueba[categorical_cols])
encoded_df_prueba_cabin = pd.DataFrame(X_prueba_encoded_cabin, columns=oh.get_feature_names_out(categorical_cols))

X_prueba = X_prueba.drop(columns=categorical_cols).reset_index(drop=True)
X_prueba = pd.concat([X_prueba, encoded_df_prueba_cabin], axis=1)

X_prueba
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import numpy as np

parameters = {
  'n_estimators': np.arange(5,51,1),
  'max_depth': np.arange(1,30,1),
  'criterion': ('gini', 'entropy')
}

rf = RandomForestClassifier(random_state=42)
gs = GridSearchCV(rf, parameters, verbose=1, n_jobs=-1, cv=3)
gs.fit(X, y)

pd.DataFrame(gs.cv_results_).sort_values('rank_test_score')
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd

# Definir el espacio de búsqueda ajustado
params = {
    'n_estimators': [100, 150, 200],  # Exploramos una cantidad moderada de árboles
    'learning_rate': [0.05, 0.1],  # Valores típicos para la tasa de aprendizaje
    'max_depth': [3, 5, 7],  # Profundidad de árboles razonable
    'min_samples_split': [2, 5, 10],  # Valores moderados de división mínima
    'min_samples_leaf': [1, 2, 4],  # Tamaño moderado de hojas mínimas
    'subsample': [0.7, 0.8, 0.9, 1],  # Exploramos fracciones del 70% al 100%
    'max_features': ['log2', 'sqrt'],  # Características a considerar en cada árbol
    'loss': ['exponential'],  # Funciones de pérdida típicas
}

# Inicializar el modelo
gb = GradientBoostingClassifier(random_state=42)

# GridSearchCV con un espacio de búsqueda moderado
gs = GridSearchCV(
    gb, params, cv=3, verbose=1, n_jobs=-1, scoring='accuracy', 
    return_train_score=True
)

# Entrenar el modelo con los parámetros
gs.fit(X, y)

# Mostrar los mejores resultados
pd.DataFrame(gs.cv_results_).sort_values('rank_test_score')
```

```python
best_gb = GradientBoostingClassifier(learning_rate=0.05, 
    loss='exponential', 
    max_depth=7, 
    max_features='log2', 
    min_samples_leaf=1, 
    min_samples_split=10, 
    n_estimators=100, 
    subsample=0.8, 
    random_state=43)
best_gb.fit(X,y)
```

```python
y_pred = best_gb.predict(X_prueba)
db_test['y_pred'] = y_pred
solution = db_test[['Id','y_pred']]
solution.to_csv('jaime12.csv', index=False)
```
